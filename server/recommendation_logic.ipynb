## Clean text data

Process text-based columns like 'name', 'ingredients', 'tags', and 'description' by removing punctuation, converting to lowercase, and potentially using techniques like stemming or lemmatization to standardize words.


**Reasoning**:
Convert relevant text columns to lowercase and remove punctuation.



```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('stopwords')
nltk.download('punkt')

def clean_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    return text

for col in ['name', 'ingredients', 'tags', 'description']:
    image_dataset[col] = image_dataset[col].apply(clean_text)

display(image_dataset[['name', 'ingredients', 'tags', 'description']].head())
```

**Reasoning**:
Apply stemming to the cleaned text columns to standardize words.



```python
stemmer = PorterStemmer()

def stem_text(text):
    # Handle non-string data by converting to string
    if not isinstance(text, str):
        text = str(text)
    return ' '.join([stemmer.stem(word) for word in text.split()])

for col in ['name', 'ingredients', 'tags', 'description']:
    image_dataset[col] = image_dataset[col].apply(stem_text)

display(image_dataset[['name', 'ingredients', 'tags', 'description']].head())
```

```python
import ast

def parse_tags(tag_string):
    try:
        return ast.literal_eval(tag_string)
    except (ValueError, SyntaxError):
        return [] # Return an empty list for unparseable strings

df['tags'] = df['tags'].apply(parse_tags)
display(df['tags'].head())
```

**Reasoning**:
Clean each tag in the 'tags' column by converting to lowercase, removing punctuation, and then rejoin them into a single string. Also, create a 'pretty_tags' column with tags joined by a comma and space for display.



```python
import re

def clean_and_join_tags(tag_list):
    cleaned_tags = []
    if isinstance(tag_list, list):
        for tag in tag_list:
            if isinstance(tag, str):
                # Convert to lowercase and remove punctuation
                cleaned_tag = re.sub(r'[^\w\s]', '', tag).lower()
                cleaned_tags.append(cleaned_tag)
    # Join cleaned tags into a single string
    return ' '.join(cleaned_tags)

def format_pretty_tags(tag_list):
    pretty_tags = []
    if isinstance(tag_list, list):
        for tag in tag_list:
            if isinstance(tag, str):
                 # Just strip leading/trailing spaces for pretty display
                pretty_tags.append(tag.strip())
    # Join tags with comma and space
    return ', '.join(pretty_tags)


df['tags'] = df['tags'].apply(clean_and_join_tags)
df['pretty_tags'] = df['tags'].apply(lambda x: format_pretty_tags(x.split())) # Apply to the cleaned string
display(df[['title', 'description', 'tags', 'pretty_tags']].head())
```

## Parse 'nutrition' column

### Subtask:
Convert the string representation of a list in the 'nutrition' column into actual Python lists of numbers. Handle any parsing errors by dropping the problematic rows.


**Reasoning**:
Convert the string representation of a list in the 'nutrition' column into actual Python lists of numbers using ast.literal_eval and drop rows with parsing errors.



```python
import ast

def parse_nutrition(nutrition_string):
    try:
        # Safely evaluate the string as a Python literal
        nutrition_list = ast.literal_eval(nutrition_string)
        # Ensure the result is a list and contains numbers
        if isinstance(nutrition_list, list) and all(isinstance(x, (int, float)) for x in nutrition_list):
            return nutrition_list
        else:
            print(f"Warning: Unexpected format for nutrition data: {nutrition_string}")
            return None
    except (ValueError, SyntaxError):
        print(f"Warning: Could not parse nutrition data: {nutrition_string}")
        return None

# Apply the function to the 'nutrition' column
df['nutrition'] = df['nutrition'].apply(parse_nutrition)

# Drop rows where parsing failed (nutrition is None)
df.dropna(subset=['nutrition'], inplace=True)

# Display the first few rows of the cleaned 'nutrition' column
display(df['nutrition'].head())
```

**Reasoning**:
The previous step successfully parsed the nutrition column. Now, extract the relevant nutrition values (calories and protein) from the lists in the 'nutrition' column and create a new column for calories and protein daily value percentage, handling cases where the list might not have enough elements.



```python
# Extract calories (first element) and protein (fourth element)
def extract_nutrition_values(nutrition_list):
    if isinstance(nutrition_list, list) and len(nutrition_list) > 3:
        # Calories is the first element (index 0)
        calories = nutrition_list[0]
        # Protein is the fourth element (index 3), assuming it's in grams
        protein_grams = nutrition_list[3]
        # Convert protein grams to Daily Value percentage (DV for protein is 50g)
        # Ensure no division by zero if DV is ever 0, although unlikely for protein DV
        protein_dv = (protein_grams / 50) * 100 if 50 > 0 else 0
        return pd.Series([calories, protein_dv])
    else:
        return pd.Series([None, None]) # Return None for both if list is invalid or too short

# Apply the extraction function and create new columns
df[['calories', 'protein_dv']] = df['nutrition'].apply(extract_nutrition_values)

# Drop rows where extraction failed (if any)
df.dropna(subset=['calories', 'protein_dv'], inplace=True)

# Display the first few rows of the new columns
display(df[['calories', 'protein_dv']].head())
```

**Reasoning**:
Create health labels based on the extracted nutrition values and combine them with existing pretty tags into a new badge_labels column. Then display the requested columns to complete the subtask.



```python
# Create health labels based on calories and protein_dv
def create_health_labels(row):
    labels = []
    if row['calories'] is not None and row['calories'] < 200:
        labels.append('low calorie')
    if row['protein_dv'] is not None and row['protein_dv'] > 20:
        labels.append('high protein')
    return labels

df['health_labels'] = df.apply(create_health_labels, axis=1)

# Combine health labels and pretty_tags into badge_labels
def combine_labels(row):
    # Ensure both are lists before concatenating
    all_labels = []
    if isinstance(row['health_labels'], list):
        all_labels.extend(row['health_labels'])
    # Split pretty_tags string into a list and extend
    if isinstance(row['pretty_tags'], str):
         all_labels.extend([tag.strip() for tag in row['pretty_tags'].split(',') if tag.strip()])
    # Remove duplicates and join with comma and space
    return ', '.join(sorted(list(set(all_labels))))

df['badge_labels'] = df.apply(combine_labels, axis=1)

# Display the first 5 rows of the specified columns
display(df[['title', 'calories', 'protein_dv', 'badge_labels']].head())
```

```python
from IPython.display import display, HTML

def show_recommendations(query=None, max_calories=None, top_n=3):
    """
    Displays recipe recommendations with images using HTML.

    Args:
        query (str, optional): A search query string. Defaults to None.
        max_calories (int, optional): The maximum calorie limit. Defaults to None.
        top_n (int, optional): The number of top recommendations to return. Defaults to 3.
    """
    recs = recommend_recipes(query=query,
                            max_calories=max_calories,
                            top_n=top_n)

    if recs.empty:
        print("No recommendations found for your criteria.")
        return

    for _, row in recs.iterrows():
        title = row["title"]
        cal = row["calories"]
        protein = row["protein_dv"]
        # Ensure badge_labels is treated as a list for joining
        badges = ", ".join(row["badge_labels"]) if isinstance(row["badge_labels"], list) else ""
        img = row["image_url"]
        url = row["url"]

        html = f"""
        <div style='border:1px solid #ccc; padding:10px; margin:10px 0;'>
          <h3>{title}</h3>
          <p>{cal:.0f} cal | Protein DV: {protein:.0f}%</p>
          <p><b>Tags:</b> {badges}</p>
          <img src="{img}" alt="{title}" style="max-width:250px; height:auto;">
          <p><a href="{url}" target="_blank">View recipe</a></p>
        </div>
        """
        display(HTML(html))

# Call the function a few times
print("Recommendations for 'chicken' with max 300 calories:")
show_recommendations(query="chicken", max_calories=300, top_n=3)

print("\nRecommendations for 'vegetarian':")
show_recommendations(query="vegetarian", top_n=3)
```
